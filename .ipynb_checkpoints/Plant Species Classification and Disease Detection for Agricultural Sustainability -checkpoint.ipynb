# Plant Species Classification and Disease Detection for Agricultural Sustainability 

## Project Objectives:

1. **Accurately Classify Plant Species from Images**:
   - Develop a deep learning model that can accurately identify and classify different plant species based on images.
   - Ensure that the model can handle a variety of plant species commonly found in agriculture.
   - Use appropriate metrics such as accuracy, F1-score, precision, and recall to measure the model's classification performance.

2. **Detect Diseases or Health Issues in Plants**:
   - Create a deep learning model capable of detecting diseases, pests, or health issues affecting plants by analyzing images.
   - The model should provide early and accurate detection to prevent the spread of diseases and optimize crop management.
   - Use relevant evaluation metrics like true positive rate, false positive rate, and overall detection accuracy to assess disease detection performance.

3. **Increase Agricultural Sustainability by Reducing Manual Inspection Efforts**:
   - Reduce the need for manual plant inspection and diagnosis by automating the process through machine learning.
   - Enhance agricultural sustainability by saving time and resources.
   - Measure the success of this objective by tracking the reduction in manual inspection hours and associated costs.

**Success Metrics**:

To measure the success of the project, you can use a combination of the following metrics:

1. **Classification Metrics**:
   - Accuracy: The percentage of correctly classified plant species.
   - F1-score: A harmonic mean of precision and recall, useful when dealing with imbalanced classes.
   - Precision: The proportion of true positive predictions among all positive predictions.
   - Recall: The proportion of true positive predictions among all actual positives.

2. **Disease Detection Metrics**:
   - True Positive Rate (Sensitivity or Recall): The percentage of actual diseases correctly detected.
   - False Positive Rate: The percentage of healthy plants incorrectly classified as diseased.
   - Overall Detection Accuracy: The accuracy of disease detection across all plant samples.

3. **Efficiency Metrics**:
   - Reduction in Manual Inspection Hours: Measure the decrease in manual inspection time achieved by the automated system.
   - Cost Savings: Assess the cost reduction associated with the reduced need for manual labor.

## Data Exploration:
This is what we do:
   - Load the dataset into your preferred data analysis tool, such as Python with libraries like Pandas and Matplotlib.
   - Explore the dataset to understand its contents, including the number of classes (plant species), the distribution of images per class, and the presence of any labels related to disease detection.
   - Visualize some sample images to get an idea of the data.

```python
import os
import matplotlib.pyplot as plt
import cv2

# Define the path to the dataset folder
dataset_path = r'C:\Users\ivy\Documents\GitHub\Capstone--Lazarus\dataset'

# Initialize dictionaries to store class information
classes = {}

# Loop through the subfolders in the dataset directory
for root, dirs, files in os.walk(dataset_path):
    for subdir in dirs:
        # Define the path to the current subfolder
        subfolder_path = os.path.join(root, subdir)
        
        # Count the number of images in the subfolder
        num_images = len(os.listdir(subfolder_path))
        
        # Store the class name (subfolder name) and the number of images in the dictionary
        classes[subdir] = num_images

# Print the class names and the number of images per class
for class_name, num_images in classes.items():
    print(f'Class: {class_name}, Number of Images: {num_images}')

# Visualize some sample images from a specific class (e.g., "maize/healthy")
sample_class = "maize/healthy"
sample_path = os.path.join(dataset_path, sample_class)

# Plot sample images from the selected class
fig = plt.figure(figsize=(10, 6))
for i, filename in enumerate(os.listdir(sample_path)[:5]):  # Plot the first 5 images
    img = cv2.imread(os.path.join(sample_path, filename))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    fig.add_subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.title(f'Sample {i + 1}')
    plt.axis('off')

plt.show()
```
